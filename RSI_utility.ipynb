```python
import yfinance as yf
import pandas as pd
```

```python
df = yf.download('SPY')
```

```python
def rsi(df, window):
    
    delta = df['Close'].diff()

    gain = delta.where(delta > 0, 0).rolling(window=window).mean()
    loss = delta.where(delta < 0, 0).rolling(window=window).mean()

    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 +rs))

    return df
```

```python
def strategy(df, overbought, oversold):

    df['Strategy'] = 0
    df.loc[df['RSI'] > overbought, 'Strategy'] = 1
    df.loc[df['RSI'] < oversold, 'Strategy'] = -1

    return df
    
```

```python
def test(df):
    df['Delta'] = df['Close'].pct_change()
    df['Returns'] = df['Delta'] * df['Strategy'].shift(1)

    df['Strat Cumulative Returns'] = ((1 + df['Returns']).cumprod() - 1) * 100
    df['SPY Cumulative Returns'] = ((1 + df['Delta']).cumprod() - 1) * 100

    return df
```

```python
def main():
    rsi(df, 14)
    strategy(df, 80, 20)
    test(df)

    return df

main()
```
